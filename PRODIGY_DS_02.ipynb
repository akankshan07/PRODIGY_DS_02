# **Exploratory Data Analysis**(EDA)

SURVIVAL PASSENGERS

```python
trainData.Survived.value_counts().plot.pie(autopct='%1.1f%%', shadow=True)
plt.title('Survival Pie Chart')
plt.figure(figsize=(2, 2))
plt.show()
```

SURVIVAL IN GENDER

```python
sns.countplot(data=trainData, x='Sex', hue='Survived')
plt.title('Survival in Gender')
plt.figure(figsize=(3, 2))
```

SURVIVAL PASSENGER BY AGE,PARCH & FARE


```python
# survival and Age with gender
sns.swarmplot(x='Sex', y="Age", hue="Survived", data=trainData)
plt.figure(figsize=(4, 3))
```

```python
# Parch & Fare with gender
for col in ['Parch', 'Fare']:
    fig, ax = plt.subplots(figsize=(4, 4))
    trainData[col].hist(ax=ax)
    plt.title(col)
    plt.show()
```

SURVIVAL AND EMBARKATION

```python
sns.countplot(data=trainData, x='Embarked', hue='Survived')
```

```python
sns.countplot(data=trainData, x='Pclass', hue='Survived')
```

```python
sns.countplot(data=trainData, x='Embarked', hue='Pclass')
```

```python
sns.countplot(data=trainData, x='SibSp', hue='Survived')
```

```python
sns.boxplot(data=trainData, x='Pclass', y='Fare')
```

```python
sns.histplot(data=trainData, x='Age', hue='Survived')
```

CORRELATION MATRIX

```python
correlation_matrix = trainData.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
